### **IMPORT THE LIBRARY**

```python
import cv2
from google.colab.patches import cv2_imshow
```

### **READ THE IMAGE**

```python
#imread() is used to read the image.
image = cv2.imread('doggo.jpg')
```

```python
#Display the Original Image
cv2_imshow(image)
```

### **GENERATE THE GRAYSCALE VERSION OF THE ORIGINAL IMAGE**

```python
''' Here we are going to use cvtColor() method to convert the original image from rgb colorspace to COLOR_BGR2GRAY which will give
us the grayscale version of the image.'''

gray_scale = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# Display the image
cv2_imshow(gray_scale)
```

### **INVERT THE IMAGE**

```python
# We will use bitwise_not() to invert the grayscale image. You can also use 255-grayscale_image to get the inverted version
inverted_gray = cv2.bitwise_not(gray_scale)
```

```python
# Display the image
cv2_imshow(inverted_gray)
```

### **GENERATE BLURRED IMAGE**

```python
# We need to use Gaussian function to generate a blurred version of the inverted image.
blurred_img = cv2.GaussianBlur(inverted_gray,(21,21),0)
inverted_blurred_img=255-blurred_img
```

```python
# Display the image
cv2_imshow(inverted_blurred_img)
```

### **CREATE THE PENCIL SKETCH**

```python
''' To obtain the pencil sketch we need to mix the grayscale and the blurred inverted images.
This can be done by dividing the grayscale withthe inverted blurred image.'''

pencil_sketch = cv2.divide(gray_scale,inverted_blurred_img,scale=256.0)
```

```python
# Display the image
cv2_imshow(pencil_sketch)
```
